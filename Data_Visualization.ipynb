```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
columns = iris.feature_names
target_names = iris.target_names
target = iris.target
```

```python
df = pd.DataFrame(data, columns=columns)
```

```python
df
```

###Bar chart

```python
plt.figure(figsize=(8, 6))
df.mean().plot(kind='bar')
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.title('Mean Value of Iris Features')
plt.show()
```

###Pie Chart

```python
counts = [len(target[target == i]) for i in range(len(target_names))] # Count the occurrences of each flower variant

plt.figure(figsize=(8, 6))
plt.pie(counts, labels=target_names, autopct='%1.1f%%')
plt.title('Distribution of Iris Flower Variants')
plt.show()
```

###Histogram

```python
plt.figure(figsize=(8, 6))
df.hist()
plt.suptitle('Histograms of Iris Features')
plt.show()
```

###Heat Map

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

###Scatter Plot


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=df, hue=target, palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

###Box Plot

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Iris Features')
plt.show()
```
